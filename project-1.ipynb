```python
import finnhub
import pandas as pd
import time
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker
from datetime import date, datetime, timedelta

# Setup finnhub client - https://finnhub.io/docs/api/
finnhub_client = finnhub.Client(api_key='cd4oqiaad3i98jhu2pggcd4oqiaad3i98jhu2ph0')

today_date = date.today()
from_unix_time = int(time.mktime((today_date - timedelta(weeks = 52)).timetuple()))
to_unix_time = int(time.mktime(today_date.timetuple()))
```

```python
# Retrieve Stock Candlestick Data
def retrieve_candlestick_data_frame(ticker_symbol):
    candle_response = finnhub_client.stock_candles(symbol=ticker_symbol, resolution='D', _from=from_unix_time, to=to_unix_time)

    candles = pd.json_normalize(candle_response)[['t', 'o', 'c', 'h', 'l', 'v']]

    candles_df = pd.DataFrame(columns=['date','unix_time', 'open', 'close', 'high', 'low', 'volume'])
    candles_df['date'] = pd.to_datetime(candles['t'][0],unit='s').astype(str)
    candles_df['year'] = pd.to_datetime(candles['t'][0],unit='s').year
    candles_df['unix_time'] = candles['t'][0]
    candles_df['open'] = candles['o'][0]
    candles_df['close'] = candles['c'][0]
    candles_df['high'] = candles['h'][0]
    candles_df['low'] = candles['l'][0]
    candles_df['volume'] = candles['v'][0]
    
    return candles_df
```

```python
# Display Stock Price Comparison

comcast_candles_df = retrieve_candlestick_data_frame('CMCSA')
verizon_candles_df = retrieve_candlestick_data_frame('VZ')

# set figure and subplot size to 4 x 4, start from 0,0 to row 3
fig = plt.figure(figsize=(16,14),dpi=100)
line_plot = plt.subplot2grid((4,4), (0, 0), rowspan=3, colspan=4)

line_plot.plot(comcast_candles_df['date'], comcast_candles_df['open'], label='Comcast', c='b')
line_plot.plot(verizon_candles_df['date'], verizon_candles_df['open'], label='Verizon', c='r')
line_plot.grid(True)
line_plot.set_title("Stock Prices")
line_plot.set_ylabel("Dollars")
line_plot.xaxis.set_major_locator(mticker.MultipleLocator(5))

plt.legend(loc=2) # set legend to upper left

for tick in line_plot.get_xticklabels():
    tick.set_rotation(45)
```

```python

# Retrieve Company Earnings
def retrieve_company_earnings(ticker_symbol):
    earnings_response = finnhub_client.company_earnings(ticker_symbol)
    earnings_df = pd.json_normalize(earnings_response).sort_values('period')

    return earnings_df
```

```python
# Display Earnings Comparison

comcast_earnings_df = retrieve_company_earnings('CMCSA')
verizon_earnings_df = retrieve_company_earnings('VZ')

# set figure and subplot size to 4 x 4, start from 0,0 to row 3
fig = plt.figure(figsize=(7,6),dpi=100)
scatter_plot = plt.subplot2grid((4,4), (0, 0), rowspan=3, colspan=4)

scatter_plot.scatter(comcast_earnings_df['period'], comcast_earnings_df['estimate'], label='Comcast Estimate', c='skyblue')
scatter_plot.scatter(comcast_earnings_df['period'], comcast_earnings_df['actual'], label='Comcast Actual', c='b')
scatter_plot.scatter(verizon_earnings_df['period'], verizon_earnings_df['estimate'], label='Verizon Estimate', c='salmon')
scatter_plot.scatter(verizon_earnings_df['period'], verizon_earnings_df['actual'], label='Verizon Actual', c='r')
scatter_plot.grid(True)
scatter_plot.set_title("Company Earnings")
scatter_plot.set_ylabel("Quarterly EPS")

plt.legend(bbox_to_anchor=(1.05, 1.0), loc='upper left')
plt.tight_layout()
```

```python
# Retrieve Social Sentiment
renamed_social_columns = {"mention": "mention_twitter", "positiveScore": "positiveScore_twitter","negativeScore": "negativeScore_twitter",
    "positiveMention": "positiveMention_twitter","negativeMention": "negativeMention_twitter","score": "score_twitter"}

def retrieve_social_sentiment_data_frame(ticker_symbol):
    social_response = finnhub_client.stock_social_sentiment(ticker_symbol)
    return pd.json_normalize(social_response, record_path='twitter').fillna(0).rename(columns=renamed_social_columns).sort_values('atTime')
```

```python
comcast_social_df = retrieve_social_sentiment_data_frame('CMCSA')
verizon_social_df = retrieve_social_sentiment_data_frame('VZ')

fig = plt.figure(figsize=(15,7),dpi=100)
social_plot = plt.subplot2grid((4,4), (0, 0), rowspan=3, colspan=4)

social_plot.plot(verizon_social_df['atTime'], verizon_social_df['positiveMention_twitter'] - verizon_social_df['negativeMention_twitter'], label='Verizon', c='r')
social_plot.plot(comcast_social_df['atTime'], comcast_social_df['positiveMention_twitter'] - comcast_social_df['negativeMention_twitter'], label='Comcast', c='b')

social_plot.grid(True)
social_plot.set_title("Social Media Sentiment")
social_plot.set_ylabel("Mentions: (Positive - Negative)")

plt.legend(loc=2)
plt.tight_layout()

for tick in social_plot.get_xticklabels():
    tick.set_rotation(45)
```

```python
# insider sentiment data - data gathered whenever a stakeholder purchases or sells their stocks.

insider_response = finnhub_client.stock_insider_sentiment('CMCSA', "2010-01-01", "2023-12-31")
insider_df = pd.json_normalize(insider_response, record_path='data').sort_values(by=['year','month'])

insider_df["date"] = insider_df['year'].astype(str) + "-" + insider_df["month"].astype(str)

fig = plt.figure(figsize=(15,8),dpi=100)
insider_plot = plt.subplot2grid((4,4), (0, 0), rowspan=3, colspan=4)
insider_plot.plot(insider_df['date'], insider_df['mspr'], label='Comcast MSPR', c='b')

insider_plot.grid(True)
insider_plot.set_title("Insider Sentiment")
insider_plot.set_ylabel("Monthly Share Purchase Ratio (MSPR)")
insider_plot.xaxis.set_major_locator(mticker.MultipleLocator(2))

plt.legend(loc=2)
plt.tight_layout()

for tick in insider_plot.get_xticklabels():
    tick.set_rotation(45)
```

```python
# analyst recommendation trends
trends = finnhub_client.recommendation_trends(symbol='CMCSA')
trends_df = pd.json_normalize(trends).sort_values('period').fillna(0)

fig = plt.figure(figsize=(6,4),dpi=100)
trend_plot = plt.subplot2grid((4,4), (0, 0), rowspan=3, colspan=4)

trend_plot.scatter(trends_df['period'], trends_df['strongBuy'], s=trends_df['strongBuy']*5, label='Strong Buy', c='b')
trend_plot.scatter(trends_df['period'], trends_df['buy'], s=trends_df['buy']*5, label='Buy', c='skyblue')
trend_plot.scatter(trends_df['period'], trends_df['hold'], s=trends_df['hold']*5, label='Hold', c='g')
trend_plot.scatter(trends_df['period'], trends_df['sell'], s=trends_df['sell']*5, label='Sell', c='salmon')
trend_plot.scatter(trends_df['period'], trends_df['strongSell'], s=trends_df['strongSell']*5, label='Strong Sell', c='r')
trend_plot.grid(True)
trend_plot.set_title("Comcast Recommendation Trends")
trend_plot.set_ylabel("Recommendation Count")

plt.legend(bbox_to_anchor=(1.05, 1.0))
plt.tight_layout()
```
